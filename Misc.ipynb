```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('/Users/bob/Downloads/SpatialSense/datasets/ca/raw/dataset_gowalla_ca_ne.csv')

# 提取日期部分，例如 2010-08-19
df['Date'] = df['UTCTime'].str[:10]

# 按日期分组并统计每个日期的数量
date_counts = df.groupby('Date').size().reset_index(name='Count')

# 打印结果
print(date_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设你已经有了 date_counts DataFrame
# 例如：
# date_counts = pd.DataFrame({
#     'Date': ['2010-08-19', '2010-08-20', '2010-08-21'],
#     'Count': [120, 95, 142]
# })

# 设置画图尺寸（可选）
plt.figure(figsize=(10, 5))

# 画柱状图
plt.bar(date_counts['Date'], date_counts['Count'])

# 添加标签
plt.xlabel('Date')
plt.ylabel('Number of Records')
plt.title('Number of Records per Date')
plt.xticks(rotation=45)  # 日期标签旋转，防止重叠

# 显示图形
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_trajectory(df, uid, day):
    subset = df[(df['uid'] == uid) & (df['d'] == day)].sort_values(by='t')
    plt.figure(figsize=(8, 6))
    plt.plot(subset['x'], subset['y'], marker='o', linestyle='-', markersize=4)
    plt.title(f'Trajectory of UID {uid} on Day {day}')
    plt.xlabel('x (grid)')
    plt.ylabel('y (grid)')
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
plot_trajectory(df, uid=24999, day=3)
```

```python
import pandas as pd
import os

# 读取数据
df = pd.read_csv('/Users/bob/Downloads/Dataset2/10142719/yjmob100k-dataset1.csv')  # 替换为你的实际文件路径

# 只保留前20000条记录
df = df.head(200000)

# 创建输出文件夹
output_dir = '/Users/bob/Downloads/Dataset2/YJMob100K_dataset'
os.makedirs(output_dir, exist_ok=True)

# 遍历所有 (uid, d) 组合
for (uid, d), group in df.groupby(['uid', 'd']):
    filename = f'uid_{uid}_day_{d}.csv'
    filepath = os.path.join(output_dir, filename)
    group.sort_values(by='t').to_csv(filepath, index=False)
```
